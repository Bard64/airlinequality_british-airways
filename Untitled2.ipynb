```python
from textblob import TextBlob
texts=df['text2']
sentiments_list=[]
for text in texts:
    blob = TextBlob(text)
    # print(f'### tag = {blob.noun_phrases}')
    sentiment = blob.sentiment
    sentiments_list.append(sentiment.polarity)
    
```

```python
def replacer(data):
    data1=data.replace("['",'').replace("']",'')
    return data1 
```

```python
df['data_flown']=df['data_flown'].apply(replacer)
# df.data_flown[0].replace("['",'').replace("']",'')
```

```python
flight_reat={'Economy Class':[],'Business Class':[],'Premium Economy':[],'First Class':[]}
flight_reat_flown={'Economy Class':[],'Business Class':[],'Premium Economy':[],'First Class':[]}
class_list=list(df['Seat_Type'].unique())
class_list.remove('[]')

for className,flightC in zip(class_list,flight_reat):
    print(className)
    data=df[df['Seat_Type']==className]['text2']
    data_F=df[df['Seat_Type']==className]['data_flown']
    # print(data)
    for D,F in zip(data,data_F):
        blob=TextBlob(D)
        flight_reat[flightC].append(blob.sentiment.polarity)
        print(F)
        flight_reat_flown[flightC].append(pd.to_datetime(F,format='%B %Y',errors='coerce'))
```

```python
def plot_mean(data, classes, talk=0):
    finall = {'Economy Class': {}, 'Business Class': {}, 'Premium Economy': {}, 'First Class': {}}
    
    for Calss in classes:
        output = {}  # تعریف `output` در داخل حلقه برای هر کلاس
        for i in range(2015, 2026):
            Res = 0
            c = 0
            if talk == 1:
                print(f'res {Calss} in {i} is :')
            for j in range(len(data[data['class'] == Calss])):
                if data[data['class'] == Calss].iloc[j]['flown'].year == i:
                    c += 1
                    Res = Res + data['reat'][j]
                    if talk == 1:
                        print(f'{Res}')
            if c == 0:
                output[i] = 'nan'  # تنظیم مقدار به 'nan' به جای 0
            else:
                output[i] = Res / c
            if talk == 1:
                print('--' * 50)
                if output[i] != 'nan':
                    print(f'output is {output[i]}')
                else:
                    print(f'output is nan')
                print('**' * 50)
        finall[Calss] = output
    return finall
```

```python
def dict2df(names):
    df1=pd.DataFrame()
    for name in names:
        df={'flown':[],'rate':[],'class':[]}
        df['flown']=flight_reat_flown[name]
        df['rate']=flight_reat[name]
        df['class']=[name]*len(flight_reat[name])
        df1 = pd.concat([df1,pd.DataFrame.from_dict(df)])
    return df1
```
